```python
import netket as nk
import json
from qutip import *
import numpy as np
import time
import multiprocessing as mp
from collections import OrderedDict
from pickle import dump
import os
import matplotlib.pyplot as plt
import scipy
from matplotlib import gridspec
from functools import reduce
plt.style.use('seaborn')
from scipy.stats import norm
```

```python
#load data for both constant and variable coefficients
N10M10_con = []
with open('2021_summer_data/N10M10_con_50runs_0601.json') as f:
    for line in f:
        N10M10_con.append(json.loads(line))

N10M10_var = []
with open('2021_summer_data/N10M10_var_50runs_0601.json') as f:
    for line in f:
        N10M10_var.append(json.loads(line))
```

```python

del_e_con = np.array(N10M10_con[0])
del_e_var = np.array(N10M10_var[0])
del_e = np.stack((del_e_con,del_e_var),axis=1)
colors = ['blue', 'green']

bins = np.arange(0, 0.8, 0.01)
plt.hist(del_e, bins=bins,color = colors,label = ['Constant A','Varying A'])
plt.xlabel('Energy error')
plt.legend()
plt.show()
```

```python
del_v_con = np.array(N10M10_con[1])
del_v_var = np.array(N10M10_var[1])
del_v = np.stack((del_v_con,del_v_var),axis=1)
colors = ['blue', 'green']

bins = np.arange(0, 1.1, 0.01)
plt.hist(del_v, bins=bins,color = colors,label = ['Constant A','Varying A'])
plt.xlabel('State vector error')
plt.legend()
plt.show()
```

```python
time_con = np.array(N10M10_con[2])
time_var = np.array(N10M10_var[2])
time = np.stack((time_con,time_var),axis=1)
colors = ['blue', 'green']

bins = np.arange(40, 65, 1)
plt.hist(time, bins=bins,color = colors,label = ['Constant A','Varying A'])
plt.xlabel('Runtime')
plt.legend()
plt.show()
```

```python
iteration = np.linspace(0,49,num=50)
plt.plot(iteration,del_e_con,'g>',markersize=8,label='Constant A')
plt.plot(iteration,del_e_var,'b^',markersize=8,label='Varying A')
plt.ylabel('Energy error')
plt.legend()
plt.show()
```
